##### Задача 1:
Имеется колода в 52 карты. Найти число возможностей вытянуть из неё 4 карты так, чтобы среди них был хотя бы один туз.

```python
#Cобытие хотя бы 1 туз из четырех карт - это одно из возможных событий: 1 туз, 2, 3 или 4.
#Эти события не могут произойти одновременно, значит мы должны просуммировать вероятности наступления этих событий,
#умножив каждое слагаемое на число сочетаний 3,2,1,0 различных карт в колоде из 48 карт.

def combinations(n: int, k: int) -> int
    return np.math.factorial(n)/(np.math.factorial(k)*np.math.factorial(n-k))
```

```python
subtotal = 0
for i in range(1,5):
    subtotal = subtotal + combinations(4, i)*combinations(48, 4-i)
print(f"Ответ по задаче №1: {int(subtotal)}")
```

##### Задача 2:
Из 60 вопросов, входящих в экзаменационные билеты, студент знает 50. Случайным образом студент вытягивает 3 вопроса.
Какова вероятность, что все выбранные вопросы знакомы студенту?
Какова вероятность что два из трёх вопросов знакомы студенту?

```python
#2.1
#Количество способов, которыми можно вытянуть 3 билета из 60: это кол-во сочетаний 3 из 60 - это будет нашим знаменателем.
#Количество способов, которыми можно вытянуть 3 билета из тех, которые выучил студент т.е. из 50 - это кол-во способов получить
#события, благоприятные для исхода "студент знает все 3 билета" - это числитель.
total_probability = combinations(50, 3)/combinations(60, 3)
print(f"Ответ по задаче №2.1: {total_probability}")
```

```python
#2.2
#Знаменатель в задаче не изменяется.
#Числитель:
#первый множитель: Количество способов, которыми можно вытянуть 2 билета из тех, которые выучил студент т.е. из 50 - это кол-во
#способов получить события, благоприятные для исхода "студент знает 2 билета". 
#второй множитель: Количество способов, которыми можно вытянуть 1 билета из тех, которых студент не знает т.е. из 10 - это кол-во
#способов получить события, благоприятные для исхода "студент не знает 1 билет". 
total_probability = combinations(50, 2)*combinations(10, 1)/combinations(60, 3)
print(f"Ответ по задаче №2.2: {total_probability}")
```
